```python
import re
file_obj=open('sentences.txt','r').read()
lines=file_obj.split('\n')
a=[]
for line in lines:
    t=re.split('[^a-z]',line.lower().strip())
    a.append(t)
print(a)
```

```python
d=0
e=dict()
for line in a:
    while ('') in line:
        line.remove('')
    for word in line:
        if not (word in e):
            e[d]=word
            d+=1
            print (e[d-1])
print (d)
    
```
